```python
import pandas as pd

def validate_bpg_format(bpg_str):
    parts = bpg_str.split("~")
    return len(parts) == 3

def main():
    valid_rows = []
    invalid_rows = []

    with open("BPG.csv", "r", encoding="ISO-8859-1", errors="ignore") as f:
        lines = f.readlines()

    for idx, line in enumerate(lines):
        bpg = line.strip()
        if validate_bpg_format(bpg):
            valid_rows.append((idx, bpg))
        else:
            invalid_rows.append((idx, bpg))

    print(f"âœ… Valid rows: {len(valid_rows)}")
    print(f"âŒ Invalid rows: {len(invalid_rows)}")

    if invalid_rows:
        print("\nâš ï¸ Invalid row details:")
        for idx, val in invalid_rows:
            print(f"  Row {idx + 1}: {val}")

        # Save to CSV
        pd.DataFrame(invalid_rows, columns=["Row", "BPG"]).to_csv("invalid_rows.csv", index=False)
        print("\nğŸ“ Saved invalid rows to invalid_rows.csv")

if __name__ == "__main__":
    main()
```
